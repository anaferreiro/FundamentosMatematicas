```python
#Prueba G1
import sympy as sp
x=sp.symbols('x')
f=-x**2*sp.log(x)
A= sp.integrate(-f, (x,1,2))
print(A)
#Falta aproximarla por trapecio compuesto
```

```python
#Prueba G2
f=2*x**2 - 7*x +5
g=-x**2 + 8*x -7
d=sp.solve(f-g)
print(d)
from sympy.plotting import plot
%matplotlib inline
fig=plot(f,show=False, line_color='red')
fig2=plot(g,show=False, line_color='green')
fig.extend(fig2)
fig.show()
A=sp.integrate(g-f, (x,1,4))
print("Area= ", A)
#Falta aproximarla por trapecio compuesto
```
