- Serie 
$$
\displaystyle \sum \frac{(-1)^n}{(2n)!}
$$
Es una serie convergente

```python
import math
n=20
suma=0
for k in range(0,n):
    suma+= (-1)**k /math.factorial(2.*k)

print("Valor de convergencia= ", suma)
```

- Ejr 6: Determinar la serie de McLaurin y su radio de convergencia para la funci√≥n 
$$
\displaystyle f(x)= \frac{1}{(1-x)^2}
$$

```python
%matplotlib inline
import sympy as sp
x=sp.symbols('x')
f=1./(1-x)**2
fig=sp.plot(f,show=False, line_color='blue',xlim=(-1,0.5),ylim=(-0.1,2))
color=['red','green', 'gray', 'pink']
i=1
for k in range(3,20,4):
    pk=sp.series(f,x,0,k)
    pk=pk.removeO()
    figk=sp.plot(pk,show=False,line_color = f'C{i}')
    fig.extend(figk)
    i+=1
fig.show()
```

Ejr 8: Calcular la serie de Taylor centrada en $x_0=1$ asociada a la funci√≥n 
$$
\displaystyle f(x) = \frac{1}{x} 
$$

```python
%matplotlib inline
import sympy as sp
x=sp.symbols('x')
f=1./(4-2*x)
fig=sp.plot(f,show=False, line_color='blue',xlim=(0.5,1.5),ylim=(0.3,1.1))
x0=1
i=1
for k in range(3,10,4):
    pk=sp.series(f,x,x0,k)
    pk=pk.removeO()
    figk=sp.plot(pk,show=False,line_color = f'C{i}')
    fig.extend(figk)
    i+=1
fig.show()
```
