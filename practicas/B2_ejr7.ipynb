Ejr 7: Calcular el volumen del sólido de revolución (elipsoide) generado por la rotación alrededor del eje OX, de la región plana encerrada dentro de la elipse de ecuación
$$
(x-3)^2+3y^2=4
$$
Calcular también el área superficial del elipsoide (no es capaz de hacerla).

En la ecuación implícita de la elipse, se puede despejar la variable $y$ en función de la variable $x$.
$$
\displaystyle y=\pm \sqrt{\frac{4-(x-3)^2}{3}}
$$
El sólido mendionado se obtiene por rotación alrededor del eje OX de la parte positiva,
$$
\displaystyle y= \sqrt{\frac{4-(x-3)^2}{3}}
$$


```python
import sympy as sp
x=sp.symbols('x')
f=sp.sqrt((4-(x-3)**2)/3)
ptos=sp.solve(f,x)
print(ptos)
V=sp.pi*sp.integrate(f**2,(x,ptos[0],ptos[1]))
print("volumen=", V)
from sympy.plotting import plot
%matplotlib inline
#fig=plot(f,show=True,xlim=(0,6))
```

Ejr8: Calcular el área de la región delimitada por la función $f(x)=e^{-x}|\sin(x)|$ en el intervalo $[0,2\pi]$

```python
g=sp.exp(-x)*sp.Abs(sp.sin(x))
g1=sp.exp(-x)*sp.sin(x)
g2=-sp.exp(-x)*sp.sin(x)
%matplotlib inline
from sympy.plotting import plot
fig=plot(g,xlim=(0,8),ylim=(-0.1,0.4))
fig2=plot(g1,show=False,xlim=(0,8),ylim=(-0.1,0.4),line_color='red')
fig3=plot(g2,show=False,xlim=(0,8),ylim=(-0.1,0.4),line_color='green')
fig.extend(fig2)
fig.extend(fig3)
fig.show()
sol=sp.solve(g1,x)
print(sol)
sol=sp.solve(g2,x)
print(sol)
```

```python
A1=sp.integrate(g1,(x,0,sp.pi))
A2=sp.integrate(g2,(x,sp.pi,2*sp.pi))
print("Area= ", A1+A2)
```

Ejr13: Calcular el área comprendida entre las curvas $y=1/cos^2(x)$, $y=\sin^6(x)$ para $x\in[0,\pi/4]$

```python
g=1/sp.cos(x)**2
f=sp.sin(x)**6
fig=plot(g,line_color='blue',xlim=(0,4),ylim=(-0.1,5),show=False)
fig1=plot(f,line_color='red',show=False)
fig.extend(fig1)
fig.show()
```

```python
A=sp.integrate(g-f,(x,0,sp.pi/4))
print("Area= ",A)
```
