Ejr2: Calcular el área de la región plana delimitada por la curva de ecuación $y=\sin(4\ln(x))$, el oje Ox y las rectas vectirales $x=1$ y $x=2$.

```python
import sympy as sp
x=sp.symbols('x')
f=sp.sin(4*sp.log(x))
from sympy.plotting import plot, plot_parametric
from matplotlib import pyplot as plt
%matplotlib inline
fig=plot(f,show=False,xlim=(0.5,2.5),ylim=(-1.1,1.1))

y=sp.symbols('y')
#fig1=plot_parametric( (y,0.5),show=False,line_color='red')  #para dibujar y=0.5, recta horizontal
#fig.extend(fig1)
fig2=sp.plot_implicit(sp.Eq(x,2),show=False, line_color='green') #Para recta vertical
fig.extend(fig2)
fig2=sp.plot_implicit(sp.Eq(x,1),show=False, line_color='yellow') #Para recta
fig.extend(fig2)
fig.show()
```

```python
area=sp.integrate(f,(x,1,2))
print("area= ",area.evalf())
```

Ejr4: Calcular el área de la región plana delimitada por la curva de la ecuación 
$$
y= \displaystyle \frac{1}{x (1 - \ln(x))},
$$
el eje OX y las rectas verticales $x=3$ y $x=4$

```python
g=1./(x*(1-sp.log(x)))
fig=plot(g,show=False,xlim=(2.95,4.5),ylim=(-5,5))
fig2=sp.plot_implicit(sp.Eq(x,3),show=False, line_color='green') #Para recta vertical
fig.extend(fig2)
fig2=sp.plot_implicit(sp.Eq(x,4),show=False, line_color='yellow') #Para recta
fig.extend(fig2)
fig.show()
area=sp.integrate(-g,(x,3,4))
print("area=", area)
```
