```python
import sympy as sp
x=sp.symbols('x')
f=x**3
T=2*sp.pi
w=2*sp.pi/T
a0=2/T * sp.integrate(f,(x,-sp.pi,sp.pi))
serie=a0/2.
n=5
for k in range(1,n):
    ak=2/T* sp.integrate(f*sp.cos(k*w*x),(x,-sp.pi,sp.pi))
    bk=2/T* sp.integrate(f*sp.sin(k*w*x),(x,-sp.pi,sp.pi))
    serie+=ak*sp.cos(k*w*x)+bk*sp.sin(k*w*x)
    
serie1=serie



n=15
for k in range(1,n):
    ak=2/T* sp.integrate(f*sp.cos(k*w*x),(x,-sp.pi,sp.pi))
    bk=2/T* sp.integrate(f*sp.sin(k*w*x),(x,-sp.pi,sp.pi))
    serie+=ak*sp.cos(k*w*x)+bk*sp.sin(k*w*x)
    
serie2=serie


from sympy.plotting import plot
%matplotlib inline
plot(serie)
```

```python
import numpy as np
gserie1=sp.lambdify(x,serie1)
gseries1=np.vectorize(gserie1)

gserie2=sp.lambdify(x,serie2)
gseries2=np.vectorize(gserie2)
```

```python
#Metodo para evaluar la función periódica
def f(x):
    a=-np.pi
    b=np.pi
    interval=b-a
    s=(x - a) % interval + a
    return s**3
```

```python
y = np.linspace(-5*np.pi, 3*np.pi,100)
#y = np.linspace(-50, 30,500)



%matplotlib inline
import matplotlib.pyplot as plt 
plt.plot(y,f(y))
plt.plot(y,gseries1(y),'-.')
plt.plot(y,gseries2(y),':')
```
